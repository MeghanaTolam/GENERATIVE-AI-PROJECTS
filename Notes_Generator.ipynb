```python
from transformers import pipeline
notes_generator=pipeline(
    "text-generation",
    model="gpt2"
)
chapter=input("Enter chapter name:")
prompt_notes=f"""
write detailed study notes for the chapter:
{chapter}
Include explanation,Examples,and Key points.
"""
detailed_notes=notes_generator(
    prompt_notes,
    max_length=300,
    num_return_sequences=1
    )
print("\n DETAILED NOTES;\n")
print(detailed_notes[0]['generated_text'])
prompt_summary=f"""
summarize the chapter {chapter} into one page,
make it simple and exam_oriented
"""
summary=notes_generator(
    prompt_summary,
    max_length=200,
    num_return_sequences=1
)
print("\n ONE_PAGE SUMMARY;\n")
print(summary[0]['generated_text'])
prompt_definitions=f"""
list important definition and formulas from the chapter:
(chapter)
use bullet points
"""
definitions=notes_generator(
    prompt_definitions,
    max_length=200,
    num_return_sequences=1
)
print("\n DEFINITIONS & FORMULAS:\n")
print(definitions[0]['generated_text'])
```
