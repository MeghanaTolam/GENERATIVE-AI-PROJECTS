```python
from transformers import pipeline
import pandas as pd
sentiment = pipeline("sentiment-analysis")
reviews="""the laptop starts quickly and works smoothly for daily tasks
the screen quality is clear amd good for watching videos
battery backup is average and needs charging often
the fan makes noise when many apps are running
overall,it is a good laptop for student but not for gaming.
"""
sentences=reviews.split("\n") # Split by newline for better sentence separation
pros=[]
cons=[]
for s in sentences:
    s=s.strip()
    if s=="":
        continue
    result=sentiment(s)[0]
    if result["label"]=="POSITIVE":
        pros.append(s)
    elif result["label"]=="NEGATIVE":
        cons.append(s)

pros=pros[:2]
cons=cons[:2]

df=pd.DataFrame({
   "pros":pros,
   "cons":cons
})
df
```
